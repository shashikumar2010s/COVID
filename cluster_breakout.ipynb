```python
import numpy as np
import pandas as pd
from google.cloud import bigquery
from pandas import ExcelWriter
```

```python
%load_ext google.cloud.bigquery
client = bigquery.Client()
```

```python
sql_cluster = """select date, location_key, cumulative_confirmed, country_name,subregion1_name,subregion1_code from covid19_open_data.covid19_open_data where country_code='US' and aggregation_level=1"""
```

```python
df_cluster = client.query(sql_cluster).to_dataframe()
```

```python
# cur_dir = %pwd
# data_path = cur_dir + '/data'
# writer = ExcelWriter(data_path + "/" + "cluster_breakout.xlsx")
# df_cluster.to_excel(writer, sheet_name='Sheet1')
# writer.save()
```

```python
df_cluster.shape
```

```python
df_cluster['date'] = pd.to_datetime(df_cluster['date'], errors='coerce')
```

```python
df_cluster['week_number'] = df_cluster['date'].dt.week
```

```python
df_cluster = df_cluster.astype({"week_number":str})
df_cluster['uniq'] = df_cluster.subregion1_code + df_cluster.week_number
# df_cluster
```

```python
df_cluster['cumulative_weekly'] = df_cluster['cumulative_confirmed'].groupby(df_cluster['uniq']).transform('max')
```

```python
df_cluster_sub = df_cluster[["uniq","cumulative_weekly","week_number","subregion1_name"]]
```

```python
df_cluster_sub = df_cluster_sub.drop_duplicates()
df_cluster_sub = df_cluster_sub.fillna(0)
df_cluster_sub = df_cluster_sub.astype({"week_number":int,"cumulative_weekly":int})
```

```python
df_pivot_cluster = df_cluster_sub.pivot(index='week_number', columns=["subregion1_name"],values='cumulative_weekly')
# df_pivot_cluster.sort_values('week_number', ascending=False)
# df_pivot_cluster
```

```python
num_of_weeks = df_pivot_cluster.shape[0] - 1
```

```python

all_loc_keys = df_pivot_cluster.columns
all_nan_countries = []
df_clean_cluster = df_pivot_cluster
for key in all_loc_keys:
    if df_clean_cluster[key].isna().sum() == df_clean_cluster.shape[0]:
        all_nan_countries.append(key)
print(f"States having all NaN values in cumulative_confirmed column: {all_nan_countries}, count: {len(all_nan_countries)}")
```

```python
df_pivot_cluster = df_pivot_cluster[:df_pivot_cluster.shape[0] - 1]
```

```python
import matplotlib.pyplot as plt
# fig = plt.figure(figsize=(55,15))
# # ax = fig.add_axes([0,0,1,1])
# ax.bar(df_pivot_cluster.index, df_pivot_cluster.columns)
# plt.xticks(rotation=90, ha='right')
# plt.grid()
# plt.show()



fig = plt.figure(figsize=(25,25))
ax = fig.add_axes([0,0,1,1])
ax.plot(df_pivot_cluster)
# df_pivot_cluster.plot()
ax.legend(df_pivot_cluster.columns, loc = 'best')
```

## Insights

- From the above plot it is clear that, in New york state the breakout began earlier than all other states during 10th - 15th week of 2020
- In States like Florida and California the outbreak was seen at later weeks, but are now among the top three high infected states
- Most states have been showing linear growth in cumulative confirmed case
- Where as few states are showing exponential increase in the confirmed cases
