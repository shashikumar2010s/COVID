```python
import numpy as np
import pandas as pd
from google.cloud import bigquery
from pandas import ExcelWriter
```

```python
%load_ext google.cloud.bigquery
client = bigquery.Client()
```

```python
sql_fatalities = """select date, location_key, cumulative_deceased, country_name from covid19_open_data.covid19_open_data where location_key=country_code"""
```

```python
df_fatalities = client.query(sql_fatalities).to_dataframe()
```

```python
df_fatalities.info()
```

```python
df_pivot_fatal = df_fatalities.pivot(index='date', columns=["country_name"],values='cumulative_deceased')
```

```python
df_pivot_fatal = df_pivot_fatal[:-2]
# df_pivot_fatal
```

```python
all_loc_keys = df_pivot_fatal.columns
all_nan_countries = []
df_clean_fatal = df_pivot_fatal
for key in all_loc_keys:
    if df_clean_fatal[key].isna().sum() == df_clean_fatal.shape[0]:
        all_nan_countries.append(key)
        df_clean_fatal = df_clean_fatal.drop(columns=key)
df_clean_fatal = df_clean_fatal.fillna(0)
print(f"countries having all NaN values in confirmed_deceased column: {all_nan_countries}, count: {len(all_nan_countries)}")
```

```python
new = pd.DataFrame(df_clean_fatal.max())
```

```python
new = new.rename(columns = {0 : "total_deceased"})
# new = test_set.set_index('date')
new = new.reset_index()
```

```python
new = new.sort_values(by=['total_deceased'], ascending=False)
```
